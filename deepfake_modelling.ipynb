```python
from tensorflow.python.keras.preprocessing.image import img_to_array
import os
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img, array_to_img
from sklearn.model_selection import train_test_split
import numpy as np
X= []
Y= []

real = ('/gdrive/MyDrive/Colab Notebooks/DeepFake/Real/RealFrames')
fake= ('/gdrive/MyDrive/Colab Notebooks/DeepFake/Fake/FakeFrames')
real_data = os.listdir(real)[0:25]

fake_data = os.listdir(fake)[0:25]

for img in real_data:
  real_join = os.path.join(real,img)
  #print(real_join)
  X.append(img_to_array(load_img(real_join)).flatten()/255.0)
  Y.append(1)
for img in fake_data: 
  fake_join = os.path.join(fake,img)
  print(fake_join)
  X.append(img_to_array(load_img(fake_join)).flatten()/255.0)
  Y.append(0)

Y_or= Y

#Performing Normalization
X= np.array(X)
Y = to_categorical(Y,2)
print(X.shape)    
print(X.size)    
print(len(X))     

X= np.reshape(1440,1440,3)
#print(X)

X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size = 0.2, random_state=5)
```

```python
base_model = InceptionV3(input_shape = (250,250,3),include_top=False, weights='imagenet')
#Making all the layers non-trainable so that we can increase performance
# We use ADAM Optimizer
for layer_model in base_model.layers:
  layer_model.trainable = False
```

```python
pip list | grep tensorflow
```

```python
model = Sequential()
model.add(TimeDistributed(base_model,input_shape=(None,45,250,250,3)))
model.add(TimeDistributed(Flatten()))
model.add(LSTM(2048,dropout=0.5,return_sequences=False))
model.add(Dense(512))
model.add(Dense(1,activation="sigmoid"))
#relu helps models to overcome the vanishing gradient issue allowing them to learn faster and better performance

#model.compile(optimizer = 'Adam', loss = 'categorical_crossentropy')
```

```python
from keras.optimizers import Adam
#binary cross entropy used because each classification can be reduced to a binary choice (here Pristine or DeepFake)
model.compile('adam',loss='categorical_crossentropy')
```
